```python
import pandas as pd
import os
import glob

directory_path = '/sc-projects/sc-proj-cc12-covid_lav/S_counts'

file_list = glob.glob(os.path.join(directory_path, '*.txt'))

combined_counts = pd.DataFrame()

for file in file_list:
    counts = pd.read_csv(file, sep='\t', index_col=0, skiprows=1, usecols=[0, 6])
    sample_name = os.path.basename(file).replace('.txt', '')
    counts.columns = [sample_name]

    if combined_counts.empty:
        combined_counts = counts
    else:
        combined_counts = combined_counts.merge(counts, left_index=True, right_index=True, how='outer')

combined_counts.to_csv('combined_LAV_bulk.csv')
```

```python
viral_genes = [
    'ORF1ab', 'S', 'ORF3a', 'E', 'M', 'ORF6', 
    'ORF7a', 'ORF7b', 'ORF8', 'N', 'ORF10'
]

viral_genes = set(viral_genes)
total_viral_reads = data[data['Geneid'].isin(viral_genes)].drop('Geneid', axis=1).sum()
total_reads_per_sample = data.drop('Geneid', axis=1).sum()

viral_read_fraction = total_viral_reads / total_reads_per_sample

viral_read_fraction_df = viral_read_fraction.reset_index()
viral_read_fraction_df.columns = ['Sample', 'Viral Read Fraction']


viral_read_fraction_df['Time Point'] = viral_read_fraction_df['Sample'].str.extract(r'_(\d+)_').astype(int)

# Now, let's sort by the Time Point
sorted_viral_read_fraction_df = viral_read_fraction_df.sort_values('Time Point')

# Plotting the sorted viral read fraction
plt.figure(figsize=(10, 5))
plt.bar(sorted_viral_read_fraction_df['Sample'], sorted_viral_read_fraction_df['Viral Read Fraction'], color='tab:blue')
plt.title('Viral Read Fraction per Sample (Sorted by Time Point)')
plt.xlabel('Sample')
plt.ylabel('Viral Read Fraction')
plt.xticks(rotation=90)
plt.tight_layout()

# Save the sorted plot to a file
sorted_viral_plot_path = 'figs/sorted_viral_read_fraction_plot_WT`.png'
plt.savefig(sorted_viral_plot_path, dpi=300, bbox_inches='tight')

sorted_viral_plot_path, sorted_viral_read_fraction_df[['Sample', 'Viral Read Fraction']].reset_index(drop=True)
```

```python

total_host_reads = total_reads_per_sample - total_viral_reads


plot_data = pd.DataFrame({
    'viral_count_sum': total_viral_reads,
    'host_count_sum': total_host_reads
})

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(plot_data['viral_count_sum'], plot_data['host_count_sum'], alpha=0.6)
plt.title('Host vs Viral Count Sum')
plt.xlabel('Viral Count Sum')
plt.ylabel('Host Count Sum')
plt.xscale('log')  
plt.yscale('log') 
plt.grid(True)

scatter_plot_path = 'figs/host_vs_viral_count_scatter_plot_wt.png'
plt.savefig(scatter_plot_path, dpi=300)

scatter_plot_path
```
